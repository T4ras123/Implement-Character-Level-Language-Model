```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
a = torch.randn((64, 60)) 
W1 = torch.randn((60, 300))
hpreact = a@ W1
h = torch.tanh(hpreact)
h.shape
```

```python
plt.hist(hpreact[0, :], bins = 40)
print(a[0, :].mean(0, keepdim=True), a[0, :].std(0, keepdim=True))
print(hpreact[0, :].mean(0, keepdim=True), hpreact[0, :].std(0, keepdim=True))
```

```python
plt.figure(figsize=(20, 10))

plt.imshow(h.abs() > 0.99, cmap='grey', interpolation='nearest')
```
